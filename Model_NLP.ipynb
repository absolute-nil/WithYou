```python
data={
  "stress":{
    "examples" : [
    "tough times",
    "emotional",
    "mental",
    "stressful"
    "tensed",
    "asleep",
    "legal problems",
    ],
    "centroid" : None
  },
  "anxiety": {
    "examples" : [
        "worried",
        "anxious",
        "trouble concentrating",
        "High blood pressure",
        "increased heart rate",
        "rapid breathing",
        "restlessness",
    ],
      "centroid" : None
  },
  "suicide" : {
    "examples" : [
        "extreme",
        "don't feel like living",
        "chai",
        "Feel hopeless, worthless, agitated",
        "stressful life event",
        "the loss of a loved one",
        "substance abuse problem — alcohol and drug abuse",
        "psychiatric disorder"
        "mental disorders, substance abuse, suicide, or violence, including physical or sexual abuse",
        "an unsupportive family",
        "i want to commit suicide"
    ],   
    "centroid" : None
  },
    "healthy":{
        "examples":[
            "feeling good",
            "people love me",
            "there is hope",
            "working towards my goal",
            "living my life to the fullest",
        ],
        "centroid": None
    }

}
```

```python
def get_centroid(embed, examples):
    C = np.zeros((len(examples),embed.vector_size))
    for idx, text in enumerate(examples):
        C[idx,:] = sum_vecs(embed,text)

    centroid = np.mean(C,axis=0)
    assert centroid.shape[0] == embed.vector_size
    return centroid
```

```python
# Adding Centroid to data dictionary
for label in data.keys():
    data[label]["centroid"] = get_centroid(embed,data[label]["examples"])
```

```python
for label in data.keys():
    print(f"{label}: {data[label]['examples']}")
```

```python
def get_intent(embed,data, text):
    intents = list(data.keys())
    vec = sum_vecs(embed,text)
    scores = np.array([ np.linalg.norm(vec-data[label]["centroid"]) for label in intents])
    return intents[np.argmin(scores)]
```

```python
for text in ["drug abuse","feeling good","don't feel like living","i want to commit suicide"]:
    print(f"text : '{text}', predicted_label : '{get_intent(embed, data, text)}'")
```

```python
templates = {
        "utter_greet_evening": ["Hey there! How was your day? "],
        "utter_greet_morning": ["Good Morning! Its a new day!,How was your sleep?"],
        "utter_goodbye_happy": ["Great, I'll go now. Bye bye", "bye bye", "Goodbye!"],
        "utter_goodbye_sad": ["I am sure it's gonna be alright!, remeber i am here for you always."],
        "utter_default": ["Sorry, I didn't quite follow"],
        "utter_anxiety": ["U r anxious","links"],
        "utter_suicide": ["If u live ,you'll make big","links"],
        "utter_stress": ["Stress is a part of life","links"],
        "utter_healthy": ["Great! looks like things are going well","here are some things you should follow to continue on the path to success","links"]
    }
```

```python
response_map = {
    "stress": "utter_stress",
    "anxiety": "utter_anxiety",
    "suicide":"utter_suicide",
    "healthy":"utter_healthy",
    "default": "utter_default",
}
```

```python
import random
def get_bot_response(bot_response_map, bot_templates, intent):
    if intent not in list(response_map):
        intent = "default"
    select_template = bot_response_map[intent]
    templates = bot_templates[select_template]
    return random.choice(templates)
```

```python
user_intent = get_intent(embed, data, "increased heart rate")
get_bot_response(response_map, templates, user_intent)
```

```python
for text in ["I am good and having success"]:
    user_intent = get_intent(embed, data, text)
    bot_reply = get_bot_response(response_map, templates, user_intent)
    print(f"text : '{text}', intent: {user_intent}, bot: {bot_reply}")
```
